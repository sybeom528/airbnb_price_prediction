```python
# 데이터 처리 및 분석
import pandas as pd
import numpy as np
import warnings

# 경로 설정
import os 
import sys

# 시각화
import matplotlib.pyplot as plt
import seaborn as sns

# 모델링
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelEncoder

# 평가지표
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score

# 출력 설정
warnings.filterwarnings('ignore')
pd.set_option('display.max_columns', None)
pd.set_option('display.max_rows', 150)

# 한글 폰트 설정
import platform
if platform.system() == 'Windows':
    plt.rcParams['font.family'] = 'Malgun Gothic'
elif platform.system() == 'Darwin':
    plt.rcParams['font.family'] = 'AppleGothic'
else:
    plt.rcParams['font.family'] = 'NanumGothic'

plt.rcParams['axes.unicode_minus'] = False
plt.rcParams['figure.figsize'] = (12,6)

# 시드 설정
np.random.seed(42)
random_seed = 42

print('-' * 60)
print('라이브러리 로드 완료')
print('한글 폰트 설정 완료!')
print('-'* 60)
```

```python
os.chdir(os.path.abspath('..'))
print(os.getcwd())

df = pd.read_csv('data/df_use_modeling.csv')
df.head()
```

- id, neighbourhood_cleansed, latitude, longitude는 향후 해석시에 용이하게 활용하기 위해 남겨둠

```python
df.select_dtypes('object').columns
```

```python
df.loc[df['neighbourhood_group_cleansed'].isin(['Queens','Bronx','Staten Island']), 
    'neighbourhood_group_cleansed'] = 'Other'
```

```python
id_vals = ['id','latitude','longitude','neighbourhood_cleansed','neighbourhood_group_cleansed'
        ,'price']
label_cols = ['host_is_superhost','neighbourhood_group_cleansed','host_identity_verified']
dummy_cols = ['room_type', 'property_grouped']
```

```python
df['neighbourhood_group_cleansed'].value_counts()
```

```python
map_dict = {'Manhattan':1, 'Brooklyn':2, 'Other':3}

for col in label_cols:
    le = LabelEncoder()
    if col != 'neighbourhood_group_cleansed':
        df[col] = le.fit_transform(df[col])
    else:
        df['city'] = df[col].map(lambda x: map_dict[x])
```

```python
df = pd.get_dummies(df, columns = dummy_cols, drop_first = True)
df.head()
```

```python
missing_col = df.columns[df.isna().sum() > 0]
df[missing_col].isna().sum() / len(df)
```

나머지 변수들은 전체 데이터의 20% 이하의 결측치를 보이므로 단순 대체

host location의 경우, 정보 미기입의 의미로 4로 표기

```python
df.loc[df['host_location'].isna(),'host_location'] = 4
```

자치구 데이터로 분할

맨해튼, 브루클린, 그 외

```python
X = df.drop(columns = ['log_price'])
y = df['log_price']

# 1. 공통(전체)
X_train_all, X_test_all, y_train_all, y_test_all = train_test_split(X, y, test_size=0.2, random_state=random_seed, stratify = df['city'])

# 이상치 클리핑
q1, q3 = y_train_all.quantile([0.25, 0.75])
iqr = q3 - q1
lower, upper = q1 - 1.5 * iqr, q3 + 1.5 * iqr

train_upper_count = (y_train_all > upper).sum()
train_lower_count = (y_train_all < lower).sum()

test_upper_count = (y_test_all > upper).sum()
test_lower_count = (y_test_all < lower).sum()

print(f"[Train] 상한 초과: {train_upper_count}개 ({train_upper_count/len(y_train_all)*100:.2f}%)")
print(f"[Train] 하한 미만: {train_lower_count}개 ({train_lower_count/len(y_train_all)*100:.2f}%)")
print(f"[Test]  상한 초과: {test_upper_count}개 ({test_upper_count/len(y_test_all)*100:.2f}%)")
print(f"[Test]  하한 미만: {test_lower_count}개 ({test_lower_count/len(y_test_all)*100:.2f}%)")

y_train_all = y_train_all.clip(lower = lower)
y_test_all = y_test_all.clip(lower = lower)

train_keep_mask = y_train_all <= upper
test_keep_mask = y_test_all <= upper

X_train_all = X_train_all[train_keep_mask]
y_train_all = y_train_all[train_keep_mask]

X_test_all = X_test_all[test_keep_mask]
y_test_all = y_test_all[test_keep_mask]

def get_city_data(X_train, y_train, X_test, y_test, city_code):
    
    X_tr = X_train[X_train['city'] == city_code].copy()
    y_tr = y_train[X_train['city'] == city_code].copy()
    X_te = X_test[X_test['city'] == city_code].copy()
    y_te = y_test[X_test['city'] == city_code].copy()
    
    # 'city' 컬럼 삭제
    X_tr = X_tr.drop(columns=['city'])
    X_te = X_te.drop(columns=['city'])

    return X_tr, y_tr, X_te, y_te

# --- 1. 자치구별로 데이터 쪼개기 ---
# Manhattan (1) - mnh 유지
X_train_mnh, y_train_mnh, X_test_mnh, y_test_mnh = \
    get_city_data(X_train_all, y_train_all, X_test_all, y_test_all, 1)

# Brooklyn (2) - bkl 유지
X_train_bkl, y_train_bkl, X_test_bkl, y_test_bkl = \
    get_city_data(X_train_all, y_train_all, X_test_all, y_test_all, 2)

# Other (3) - other로 풀기
X_train_other, y_train_other, X_test_other, y_test_other = \
    get_city_data(X_train_all, y_train_all, X_test_all, y_test_all, 3)
```

각각의 데이터에 대해, 결측치 대체 및 종속변수 상한/하한선 조정

```python
def preprocess_data(X_train, y_train ,X_test, y_test, missing_cols, id_vals):
    '''
    Train 데이터를 기준으로 통계값을 계산하고 Test 데이터에 동일하게 적용
    '''
    
    # 1. 단순 대체
    for col in missing_cols[1:]:
        if X_train[col].dtype.name in ['float64', 'int64']:
            fill_val = X_train[col].median()
            X_train[col] = X_train[col].fillna(fill_val)
            X_test[col] = X_test[col].fillna(fill_val)
        else:
            fill_val = X_train[col].mode()[0]
            X_train[col] = X_train[col].fillna(fill_val)
            X_test[col] = X_test[col].fillna(fill_val)

    # 2. ID 및 위치 정보 분리 및 삭제
    train_id_vals = X_train[id_vals]
    test_id_vals = X_test[id_vals]
    
    X_train = X_train.drop(columns = id_vals)
    X_test = X_test.drop(columns = id_vals)

    return X_train, y_train, X_test, y_test, train_id_vals, test_id_vals

# 1. 맨해튼 (mnh)
X_train_mnh, y_train_mnh, X_test_mnh, y_test_mnh, ids_train_mnh, ids_test_mnh = \
    preprocess_data(X_train_mnh, y_train_mnh, X_test_mnh, y_test_mnh, missing_col, id_vals)
print(f"[Manhattan] Train: {X_train_mnh.shape}, Test: {X_test_mnh.shape}")

# 2. 브루클린 (bkl)
X_train_bkl, y_train_bkl, X_test_bkl, y_test_bkl, ids_train_bkl, ids_test_bkl = \
    preprocess_data(X_train_bkl, y_train_bkl, X_test_bkl, y_test_bkl, missing_col, id_vals)
print(f"[Brooklyn]  Train: {X_train_bkl.shape}, Test: {X_test_bkl.shape}")

# 3. 기타 (other)
X_train_other, y_train_other, X_test_other, y_test_other, ids_train_other, ids_test_other = \
    preprocess_data(X_train_other, y_train_other, X_test_other, y_test_other, missing_col, id_vals)
print(f"[Other]     Train: {X_train_other.shape}, Test: {X_test_other.shape}")

# 4. 전체 (total)
X_train_all, y_train_all, X_test_all, y_test_all, ids_train_all, ids_test_all = \
    preprocess_data(X_train_all, y_train_all, X_test_all, y_test_all, missing_col, id_vals)
print(f"[Total]     Train: {X_train_all.shape}, Test: {X_test_all.shape}")

print("모든 데이터 세트 전처리 완료 및 ID 분리 성공!")
```
